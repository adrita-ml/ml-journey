```python
#Loading a small dataset(CIFAR-10)
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train/255.0
```

```python
#Data Augmentation (teaches the model better generalization)
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range = 20, #this rotates the images slightly and teaches the model that object orientation can vary
    zoom_range = 0.2, #this zooms in/out the images and teaches the model that object size can vary
    horizontal_flip = True #this flips the image and teaches the model that left/right orientation shouldn't matter
)
```

```python
#VISUALIZE augmented images

sample_image = x_train[0].reshape((1,32,32,3))

aug_iter = datagen.flow(sample_image, batch_size=1)

plt.figure(figsize=(6,6))
for i in range(4):
  plt.subplot(2,2,i+1)
  plt.imshow(next(aug_iter)[0])
  plt.axis('off')

plt.show()
```
